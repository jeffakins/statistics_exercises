## Graphing Practice

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data for plotting
x = np.arange(-100.0, 100.0, 0.01)
y = x**3 + 4*x**2 - x**2 + 3*x + 10

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='X', ylabel='Y',
       title='Your Graph')
ax.grid()

fig.savefig("test.png")
plt.show()
```
